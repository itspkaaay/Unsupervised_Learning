```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
x,y=make_blobs(n_samples=500,n_features=2,centers=5,cluster_std=1)
```

```python
x=np.array(x)
y=np.array(y)
```

```python
plt.style.use('seaborn')
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
clusters ={
    
}
colours=["green","blue","yellow","red","black"]
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))

def init_clusters(clusters,k=5):
    for kx in range(k):
        cluster={
            'color': colours[kx],
            'points': [],
            "mean" : 10*np.random.random((x.shape[1]))
        }
        clusters[kx]= cluster
    return clusters

def kmeans(clusters,k=5):
    
    for ix in range(x.shape[0]):
        dist=[]
        
        for kx in range(k):
            dist.append(distance(x[ix],clusters[kx]['mean']))
        
        minIndex= np.argmin(dist)
        clusters[minIndex]['points'].append(x[ix])
    
    return clusters
        

def plotClusters(clusters,k=5):

    for kx in range(5):
        temp= np.array(clusters[kx]['points'])
        try:
            x_axis= temp[:,0]
            y_axis= temp[:,1]
            plt.scatter(x_axis,y_axis,marker='.',color=clusters[kx]['color'])
            
        except:
            pass
    
        plt.scatter(clusters[kx]['mean'][0],clusters[kx]['mean'][1],marker='x',
                    color=clusters[kx]['color'])
    plt.show()
```

```python
clusters= init_clusters(clusters)

```

```python
clusters= kmeans(clusters)
```

```python
for kx in range(5):
    print(np.array(clusters[kx]['points']).shape)
    
```

```python
plotClusters(clusters)
```
