```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
x,y=make_blobs(n_samples=500,n_features=2,centers=5,cluster_std=1,random_state=3)
```

```python
x=np.array(x)
y=np.array(y)
```

```python
plt.style.use('default')
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
clusters ={
    
}
colours=["green","blue","yellow","red","black"]
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))

def init_clusters(clusters,k=5):
    for kx in range(k):
        cluster={
            'color': colours[kx],
            'points': [],
            "mean" : 10*np.random.random((x.shape[1]))
        }
        clusters[kx]= cluster
    return clusters



        

```

```python
def kmeans(clusters,k=5):
    
    for ix in range(x.shape[0]):
        dist=[]
        
        for kx in range(k):
            dist.append(distance(x[ix],clusters[kx]['mean']))
        
        minIndex= np.argmin(dist)
        clusters[minIndex]['points'].append(x[ix])
    
    plotClusters(clusters)
    for kx in range(k):
        u=np.array([0.0,0.0])
        for point in clusters[kx]['points']:
            u+=np.array(point)
        
        if(len(clusters[kx]['points'])!=0):
            u/= len(clusters[kx]['points'])
            clusters[kx]['mean']=u
        else:
            pass
        print(np.array(clusters[kx]['points']).shape)
        clusters[kx]['points']=[]
        
    return clusters
        
```

```python
def plotClusters(clusters,k=5):

    for kx in range(5):
        temp= np.array(clusters[kx]['points'])
        try:
            x_axis= temp[:,0]
            y_axis= temp[:,1]
            plt.scatter(x_axis,y_axis,marker='o',color=clusters[kx]['color'])
            
        except:
            pass
    
        plt.scatter(clusters[kx]['mean'][0],clusters[kx]['mean'][1],marker='x',
                    color=clusters[kx]['color'])
    plt.show()
```

 # K-Means++ clustering using sklearn to prevent local minima problem

```python
from sklearn.cluster import KMeans
```

```python
km= KMeans(n_clusters=5)
```

```python
km.fit(x)
```

```python
centers=km.cluster_centers_
```

```python
labels=km.labels_
```

```python
print(len(labels))
```

```python
plt.style.use('default')
plt.scatter(x[:,0],x[:,1],c=labels)
plt.scatter(centers[:,0],centers[:,1],marker='*',color='red')
plt.title('Clustering on prediction')
plt.show()
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.title('Actual Clusters')
plt.show()
```

Can be noted that kmeans++ ie the inbuilt method for selection of initial cluster centers used by sklearn gives more accuarate clustering and does not get stuck at local minima condition unlike the self written kmeans algorithm
